```python
import imgaug.augmenters as iaa
import imageio
```

```python
images = [imageio.v2.imread("../data/e1.png")]
seq = iaa.CoarseDropout(0.3, size_percent=0.2, per_channel=0.5)
images_aug = seq(images=images)
imageio.v2.imwrite("../data/occluded_e1.png", images_aug[0])
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 

def add_random_occlusion(img, occlusion_ratio=0.4):
    h, w = img.shape[:2]
    occlusion_w = int(w * np.sqrt(occlusion_ratio))
    occlusion_h = int(h * np.sqrt(occlusion_ratio))
    x = np.random.randint(0, w - occlusion_w)
    y = np.random.randint(0, h - occlusion_h)
    img[y:y+occlusion_h, x:x+occlusion_w] = 0  # 黑色遮挡
    return img

ori_img = cv2.imread('../data/e1.png')
made_img = add_random_occlusion(ori_img)
plt.imshow(made_img)
plt.axis('off')
plt.show()
```
